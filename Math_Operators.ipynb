```python
# Math operations

print('Welcome to the Calculator')
print('What do you want to do? \n 1 Add \n 2 Subtraction \n 3 Multiply \n 4 Division \n 5 Percentage \n 6 Average \n 7 Print and Loss')
main_menu = int(input("Enter number : "))
#applying condition based on user's selection
if main_menu==1:    #adding operator
  print('What do you want to do?')
  print("1. Sum of n numbers \n 2 Sum of 2 numbers \n 3 Sum of 3 numbers")
  b = int(input("Enter number : "))
  #applying condition based on ADD operator options
  if b==1:
    z=0
    times = int(input('How many numbers you want to add?'))
    for i in range(1,times):
       print("Enter",i," number you want to add")
       num = int(input())    #put number of x times
       z=z+num
    print("Sum of the numbers:",z)
  elif b==2:
    two_sum=0
    first_num= int(input("Enter your first number: ")) #sum of any two numbers
    second_num= int(input("Enter your Second number: "))
    two_sum = first_num + second_num  #adding the input numbers and storing in two_sum
    print("Sum of two numbers will be: ",two_sum)
  elif b==3:
    three_sum=0
    first_num= int(input("Enter your first number: ")) #sum of any three numbers
    second_num= int(input("Enter your Second number: "))
    third_num= int(input("Enter your third number: "))
    three_sum = first_num + second_num + third_num #adding the input numbers and storing in three_sum
    print("Sum of three numbers will be: ",three_sum)
elif main_menu==2:   #subtraction operator
  print('What do you want to do?')
  print("1. Subtraction of n numbers \n 2 Subtraction of 2 numbers \n 3 Subtraction of 3 numbers")
  b = int(input("Enter number : "))
  #applying condition based on Subtraction operator options
  if b==1:
    z=0
    times = int(input('How many numbers you want to Subtraction?'))
    for i in range(1,times):
       print("Enter",i," number you want to Subtraction")
       num = int(input())    #put number of x times
       z= z-num
    print("Result:",z)
  elif b==2:
    two_sub=0
    first_num= int(input("Enter your first number: ")) #Subtraction of any two numbers
    second_num= int(input("Enter your Second number: "))
    two_sub = first_num - second_num  #Subtraction the input numbers and storing in two_sub
    print("Result: ",two_sub)
  elif b==3:
    three_sub=0
    first_num= int(input("Enter your first number: ")) #Subtraction of any three numbers
    second_num= int(input("Enter your Second number: "))
    third_num= int(input("Enter your third number: "))
    three_sub = first_num - second_num - third_num #subtracting the input numbers and storing in three_sub
    print("Result: ",three_sub)
elif main_menu==3:    #multiplication operator
  print('What do you want to do? \n 1 Multiply two numbers \n 2 Multiply n numbers')
  multiply = int(input('Enter your choice here :'))
  if multiply== 1:
    first_num = int(input())
    second_num = int(input())
    multi = first_num*second_num
    print('Multiplication of two numbers :',multi)
  elif multiply== 2:
    times=int(input('How many numbers you wanted to multiply :'))
    multi_var = 1
    for i in range(1,times+1):       #loop to call n numbers for multiplication
      print("Enter",i,"your number :")
      num=int(input())
      multi_var=multi_var*num
    print('Multiplication of all the numbers :',multi_var)
elif main_menu==4:  #Division
   print('Divide any two numbers')
   dividend = int(input('Enter your dividend here :'))
   divisor = int(input('Enter your divisor here :'))
   result = dividend / divisor
   print(result)
elif main_menu==5:   #calculate percentage
   print('What do you want to do? \n 1 Calculate percentage of the following numbers \n 2 Calculate x% of the given number')
   percent=int(input('Select an option'))
   if percent==1:  #Calculate percentage of the numbers
     value=int(input('Enter the small value'))
     total_values=int(input('Enter the larger value'))
     result=(value*100/total_values)
     print(result,'%')
   elif percent ==2: #Calculate x% of the given number
    user=int(input('Enter the number : '))
    per=int(input('What percent do you want to calculate for the given number'))
    result = (user*per)/100  #calculating percentage of the number given by the user
    print(result,'%')
elif main_menu==6: #average
  print(" Average of n numbers ")
  b = int(input("How many numbers you want to average? : "))
  z=0
  for i in range(1,b+1):
     num = int(input('Enter your number here')) # taking numbers for the average operation
     z=z+num
  print("Average of the number",z/b)
elif main_menu==7:   #calculate profit and loss
   print("Let's calculate profit and loss for the product \n Do you want to calculate \n 1 with discount \n 2 without discount")
   num = int(input('Enter your choice here :'))
   if num ==1:
     cost_price=int(input('Enter the cost price :'))
     Selling_price=int(input('Enter the selling price :'))
     Discount= int(input('Please enter the discount % :'))
     Profit = (Selling_price-((Selling_price*Discount)/100))-cost_price
     print(Profit)
   elif num==2:
    cost_price=int(input('Enter the cost price :'))
    Selling_price=int(input('Enter the selling price :'))
    Profit = Selling_price - cost_price
    if Profit > 0:
      print("Profit :",Profit)
    elif  Profit < 0:
      print("Loss :",Profit)
    else:
      print("No profit No Loss")
else:
  print("For further calculation stay tuned :)")
```
